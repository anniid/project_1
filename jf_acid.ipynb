```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import requests
```

```python
acid_df= pd.read_csv('ocean_acid.csv')
jf_df= pd.read_csv('nicernum.csv')
acid_df.head()
```

```python
tacid_jf_df= pd.DataFrame({'Year':jf_df['Year'],
                         'Jellyfish Sightings':jf_df['Jellyfish sightings'],
                         'Ocean pH':acid_df['Hawaii pH']})
acid_jf_df= tacid_jf_df.dropna()
acid_jf_df.head()
```

```python
jelly= acid_jf_df['Jellyfish Sightings']
acid= acid_jf_df['Hawaii cpH']
plt.ylabel('Jellyfish Sightings')
plt.xlabel('Ocean pH Levels')

plt.scatter(acid, jelly)
plt.title('Jellyfish Sightings vs. Ocean pH Levels')
```
