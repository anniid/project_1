```python
# Dependencies
import requests
import json
import time

url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"
#url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?fq=jellyfish&facet_field=time&facet=true&begin_date=19700101&end_date=20110531&api-key=BvGxD5OWHNo66EtooPjgOPX3roOwHI4x"

api_key = "BvGxD5OWHNo66EtooPjgOPX3roOwHI4x"

# Store a search term
query = "jellyfish"

# Search for articles published between a begin and end date
#dates = "&begin_date=19700101&end_date=20110531"

begin_date= "19700101"
end_date= "20110531"
dates = "&begin_date=" + begin_date + "&end_date=" + end_date



query_url = url + "q="  + query  + "&dates=" + dates + "&api-key=" + api_key  


response = requests.get(query_url).json()
print (json.dumps(response, indent = 4, sort_keys = True)) 
```
